```python
#Dataframe practice
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Age': [25, 30, 35, 0, 45], 'Department': ['HR', 'Finance', 'IT', 'Marketing', 'Sales'], 
        'Salary': [50000, 60000, 70000, 0, 90000]}

data
```

```python
df=pd.DataFrame(data)
df
```

```python
#loc and iloc
df.iloc[1:3]
```

```python
df.loc[1:3, ['Name', 'Salary']]
```

```python
df.iloc[1:3, :2] # rows, columns and the range for this are seperated by comma, 
#here we are selecting rows 1 and 2 and columns 0 and 1
```

```python
df[["Age","Department"]] #this will give an error as we are trying to access multiple columns using single square brackets, we need to use double square brackets
```

```python
# dropping a column
df.drop('Age', axis=1) #without inplace=True, this will return a new dataframe with the column dropped, 
#but the original dataframe will remain unchanged
```

```python
df.drop('Age', axis=1, inplace=True) #with inplace=True, the original dataframe will be modified and the column will be dropped permanently
```

```python
df.shape #gives the number of rows and columns in the dataframe
```

```python
df.info() #gives the summary of the dataframe, including the number of non-null values, data types, and memory usage
```

```python
df.describe() #gives the summary statistics of the numerical columns in the dataframe, including count, mean, standard deviation, minimum, maximum, and quartiles
```

```python
#Broadcasting in pandas

df["Salary"] = df["Salary"] * 1.1 #increasing salary by 10%
df["Salary"]
```

```python
df.rename(columns={'Department': 'Dept'}, inplace=True) #renaming a column, with inplace=True, the original dataframe will be modified and the column will be renamed permanently
df
```

```python
df["Salary"].unique() #gives the unique values in the Salary column
```

```python
df["Dept"].unique() #gives the unique values in the Dept column
```

```python
df["Dept"].value_counts() #gives the count of unique values in the dept column, this will give an error as we have renamed the column to Dept, so we need to use Dept instead of dept
```

```python
df["Promoted Salary"]= df["Salary"]*10.0
df
```

```python
# data Cleaning

df.isnull() #gives a boolean dataframe indicating the presence of null values

df.isnull().sum() #gives the count of null values in each column
```

```python
#dropping or removing null values from a dataframe

df.dropna() #this will return a new dataframe with the null values dropped, but the original dataframe will remain unchanged

df.dropna(inplace=True) #with inplace=True, the original dataframe will be modified and the null values will be dropped permanently

df
```

```python
df.dropna(how='any') #this will drop the rows where any of the values are null, this is the default behavior of dropna( )

df.dropna(how='all') #this will drop the rows where all of the values are null, this will not drop any rows as there are no rows where all values are null
df
```

```python
# Dealing with missing values

#option 1 - fill with most repeated values using fillna()
#option 2 - getting rid of them using dropna()

df.fillna(0)

df.fillna(method='ffill') #this will fill the null values with the previous non-null value, this is called forward fill

df["Age"].fillna(df["Age"].mean())

df["Salary"].fillna(df["Salary"].median())
```

```python
#fillna has 2 methods
# 1. forward fill
# 2. backward fill

df["Age"].fillna(method='ffill') #this will fill the null values with the previous non-null value, this is called forward fill
df["Dept"].fillna(method='bfill') #this will fill the null values with the next non-null value, this is called backward fill
df
```

```python
#renaming values

df["Name"].replace({'Rose','Bob'})
df

df['Name']= df['Name'].replace({'Rose':'Bob'}) #this will replace the value 'Alice' with 'Rose' in the Name column, and the original dataframe will be modified permanently
df
```

```python
#dealing with duplicates

df_dup=df[df.duplicated()]
df
```

```python
df_dup=df[df.duplicated(keep='last')] #works from bottom to top

df=df.drop_duplicates()
df
```

```python
# invalid values
#Lambda 

df["Promoted Salary"] =  df["Promoted Salary"].apply(lambda x : x/10 if x > 7000 else x)
df
```

```python
name="alice_fernandes"
df[["first_name","second_name"]]=df["name"].str.split("_")
```

```python
# apply and lambda 
def mul_age(x):
    return x*2

df["Age"]=df["Age"].apply(mul_age)
df
```

```python
df["Age"]=df["Age"].apply(lambda x: x/2)
df
```

```python
#joins in pandas

# left, right, outer, inner, merge

dept_info= {
    "Dept":["HR","IT","Product","Finance"],
    "Location": ["New York", "San", "India", "Japan"],
    "Manager": ["Laura", "Steve", "Kevin", "Radhya"]

}

df2=pd.DataFrame(dept_info)
df2
```

```python
#concat
pd.concat([df, df2])
```

```python
pd.concat([df, df2], axis=1)
```

```python
pd.merge(df,df2, on= 'Dept')
```
