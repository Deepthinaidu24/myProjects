```python
#iloc -> location based indexing
s.iloc[3]
s.iloc[[1,3,4]]
```

```python
index=['apple','banana','mango','grapes','orange']
```

```python
s.index=index
s
```

```python
#loc --> label based indexing
#in label based indexing we can use the index name to access the values, here both start and stop values are included
s.loc['banana']
```

```python
fruit_protein = {'apple': 5, 'banana': 10, 'mango': 15, 'grapes': 20, 'orange': 25}

s2=pd.Series(fruit_protein, name='protein')
s2
```

```python
# conditional selection
s2>5 #masked series 
s2[s2>5] #passing masked series to the original series to get the values
```

```python
#logical operators

(s2>5) & (s2<20)
s2[(s2>5) & (s2<20)]

s2[(s2>5) | (s2<20)]
```

```python
#modifying series values    
s2['strawberry'] = 30
s2

s2['mango'] = 18
s2
```
